```python
emp={'age_years':[10,23,22,25,20,23,24,27,25,24,33,21],'salary_thousands':[15,22,24,25,30,23,34,43,19,20,52,21]}
df=pd.DataFrame(data=emp)
df
```

```python
sal=[15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

```python
print(sum(sal)/12)
```

```python
#outliers
ser=pd.Series(sal)
ser
```

```python
q1=ser.quantile(0.25)
print('Q1',q1)
q3=ser.quantile(0.75)
print('Q3',q3)
iqr=q3-q1
print('IQR',iqr)
```

```python
upside=q3+(1.5*iqr)
upside
```

```python
lower_side=q1-(1.5*iqr)
lower_side
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')

tips.tail()
```

```python
sns.relplot(x='total_bill',y='tip',data = tips)
plt.show()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)#size='size')
plt.show()
```

```python
tips.head()
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
plt.show()
```
