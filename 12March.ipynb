```python
#define function
def salary_hike(x):
    if(x>3000):
        return np.ceil(x+(x*.30))
    else:
        return np.ceil(x*1.50)
```

```python
df=pd.DataFrame({'dept':[1,2,3,4,5],
                'emp_name':['Yogesh','Radhe','Nikki','Akshay','Lalli'],
                'salary':[2589,3625,1245,6542,3214]})
df
```

```python
df['new_salary']=df['salary'].apply(salary_hike)
df
```

```python
df['name_length']=df['emp_name'].apply(len)
df
```
